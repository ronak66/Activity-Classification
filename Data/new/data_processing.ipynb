```python
# fig, asx = plt.subplots(2,2,figsize=(20,15))
fig, asx = plt.subplots(2,1,figsize=(5,5))
l=300
df = pd.read_csv('climbing_0/Accelerometer.csv')
df = df[0:l]
asx[0].plot(
    df['Milliseconds']/1000,df['X'],'-r',
    df['Milliseconds']/1000,df['Y'],'-g',
    df['Milliseconds']/1000,df['Z'],'-b'
)
asx[0].set_title('Accelerometer Data')
asx[0].set_xlabel('Time (s)')
asx[0].set_ylabel('Value')
asx[0].legend(['x-axis','y-axis','z-axis'],loc='center left')
df = pd.read_csv('climbing_0/Gyroscope.csv')
df = df[0:l]
asx[1].plot(
    df['Milliseconds']/1000,df['X'],'-r',
    df['Milliseconds']/1000,df['Y'],'-g',
    df['Milliseconds']/1000,df['Z'],'-b'
)
asx[1].set_title('Gyroscope Data')
asx[1].set_xlabel('Time (s)')
asx[1].set_ylabel('Value')
asx[1].legend(['x-axis','y-axis','z-axis'])
fig.tight_layout()
plt.savefig('climb.png')
plt.show()

fig, asx = plt.subplots(2,1,figsize=(5,5))
df = pd.read_csv('jumping_1/Accelerometer.csv')
df = df[0:l]
asx[0].plot(
    df['Milliseconds']/1000,df['X'],'-r',
    df['Milliseconds']/1000,df['Y'],'-g',
    df['Milliseconds']/1000,df['Z'],'-b'
)
asx[0].set_title('Accelerometer Data')
asx[0].set_xlabel('Time (s)')
asx[0].set_ylabel('Value')
asx[0].legend(['x-axis','y-axis','z-axis'])
df = pd.read_csv('jumping_1/Gyroscope.csv')
df = df[0:l]
asx[1].plot(
    df['Milliseconds']/1000,df['X'],'-r',
    df['Milliseconds']/1000,df['Y'],'-g',
    df['Milliseconds']/1000,df['Z'],'-b'
)
asx[1].set_title('Gyroscope Data')
asx[1].set_xlabel('Time (s)')
asx[1].set_ylabel('Value')
asx[1].legend(['x-axis','y-axis','z-axis'])
fig.tight_layout()
plt.savefig('jump.png')
plt.show()

fig, asx = plt.subplots(2,1,figsize=(5,5))
df = pd.read_csv('running_2/Accelerometer.csv')
df = df[0:l]
asx[0].plot(
    df['Milliseconds']/1000,df['X'],'-r',
    df['Milliseconds']/1000,df['Y'],'-g',
    df['Milliseconds']/1000,df['Z'],'-b'
)
asx[0].set_title('Accelerometer Data')
asx[0].set_xlabel('Time (s)')
asx[0].set_ylabel('Value')
asx[0].legend(['x-axis','y-axis','z-axis'],loc='center left')
df = pd.read_csv('running_2/Gyroscope.csv')
df = df[0:l]
asx[1].plot(
    df['Milliseconds']/1000,df['X'],'-r',
    df['Milliseconds']/1000,df['Y'],'-g',
    df['Milliseconds']/1000,df['Z'],'-b'
)
asx[1].set_title('Gyroscope Data')
asx[1].set_xlabel('Time (s)')
asx[1].set_ylabel('Value')
asx[1].legend(['x-axis','y-axis','z-axis'])
fig.tight_layout()
plt.savefig('run.png')
plt.show()

fig, asx = plt.subplots(2,1,figsize=(5,5))
df = pd.read_csv('walking_3/Accelerometer.csv')
df = df[0:l]
asx[0].plot(
    df['Milliseconds']/1000,df['X'],'-r',
    df['Milliseconds']/1000,df['Y'],'-g',
    df['Milliseconds']/1000,df['Z'],'-b'
)
asx[0].set_title('Accelerometer Data')
asx[0].set_xlabel('Time (s)')
asx[0].set_ylabel('Value')
asx[0].legend(['x-axis','y-axis','z-axis'])
df = pd.read_csv('walking_3/Gyroscope.csv')
df = df[0:l]
asx[1].plot(
    df['Milliseconds']/1000,df['X'],'-r',
    df['Milliseconds']/1000,df['Y'],'-g',
    df['Milliseconds']/1000,df['Z'],'-b'
)
asx[1].set_title('Gyroscope Data')
asx[1].set_xlabel('Time (s)')
asx[1].set_ylabel('Value')
asx[1].legend(['x-axis','y-axis','z-axis'])
fig.tight_layout()
plt.savefig('walk.png')
plt.show()
```
